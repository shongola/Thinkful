```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Military Expenditure.csv')
df = df.fillna(0)
df = df.drop(['Indicator Name','Type','Code'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df2 = df.transpose()
df2.columns = df2.iloc[0]
df2 = df2.iloc[1:]
df2['Global_Spend'] = df2.sum(axis=1)
df2 = df2.drop(df2.columns[0:264], axis=1)
df2 = df2.drop(df2.index[0:38])
df2.head()
```

```python
df2 = df2['Global_Spend'].div(1000000000).round(2)
global_spend = pd.DataFrame(data=df2)
#global_spend.reset_index(inplace = True, drop = False) 
#global_spend.columns.values[0] = "Year"
global_spend.head()
```

```python
hist = global_spend.plot.bar(figsize=(10,5))
plt.title('Global Military Expenditure')
plt.xlabel('Year')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
box = plt.boxplot(global_spend['Global_Spend'])
plt.title('Global Military Expenditure 1998-2018')
plt.xlabel('Military Expenditure')
plt.ylabel('Spending in $ Billions')
plt.show()
```

```python
usa_v_rus = df.transpose()
usa_v_rus.columns = usa_v_rus.iloc[0]
usa_v_rus = usa_v_rus.iloc[1:]
usa_v_rus = usa_v_rus.drop(usa_v_rus.columns[0:200], axis=1)
usa_v_rus = usa_v_rus.drop(usa_v_rus.columns[1:49], axis=1)
usa_v_rus = usa_v_rus.drop(usa_v_rus.columns[2:], axis=1)
usa_v_rus = usa_v_rus.div(1000000000, axis='columns').round(2)
usa_v_rus.reset_index(level=0, inplace=True)
usa_v_rus.columns.values[0] = "Year"
usa_v_rus = usa_v_rus.drop(usa_v_rus.index[0:38])
usa_v_rus.head()
```

```python
usa_v_rus_clean = usa_v_rus.melt(id_vars='Year')
usa_v_rus_clean.head()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='Year', y='value', hue='Name', data=usa_v_rus_clean)
plt.title('USA v Russia Military Spend 1998-2018')
plt.xlabel('Year')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
usa_v_rus_clean['value'] = usa_v_rus_clean['value'].astype(float)
plt.figure(figsize=(7,7))
sns.boxplot(x='Name', y='value', data=usa_v_rus_clean)
plt.title('Distribution of USA v Russian Military Spend 1998-2018')
plt.xlabel('Country')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.lmplot(x='value', y='value', hue='Name', data=usa_v_rus_clean)
plt.title('USA v Russian Military Spend 1998-2018')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
region = pd.read_csv('Military Expenditure.csv')
region = region.fillna(0)
region = region.drop(region.columns[4:62], axis=1)
region.loc[:, '2018_spend_billions'] = region.loc[:, '2018'].div(100000000).round(2)
region = region.drop(['2018'], axis=1)
region = region[region.Type != 'Country']
region.head()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(11,6))
sns.boxplot(x='Type', y='2018_spend_billions', data=region)
plt.title('Distribution of Regional Military Spend 2018')
plt.xlabel('Region Type')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x='Type', y='2018_spend_billions', data=region)
plt.title('Distribution of Regional Military Spend 2018')
plt.xlabel('Region Type')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='Type', y='2018_spend_billions', data=region, inner=None)
sns.swarmplot(x='Type', y='2018_spend_billions', data=region, color='k', alpha=0.7)
plt.title('Distribution of Regional Military Spend 2018')
plt.xlabel('Region Type')
plt.ylabel('Spending in $US Billions')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x='Type', y='2018_spend_billions', hue=None, data=region, orient='v')
#plt.legend(bbox_to_anchor=(1.05, 1))
plt.title('Regional Military Spend 2018')
plt.xlabel('Region Type')
plt.ylabel('Spending in $US Billions')
plt.show()
```
